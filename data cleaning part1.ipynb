# Skiping the rows where null values are present

```python
df=pd.read_csv('air_traffic_landings_statistics_1.csv',na_values="-",dtype={'total_landed_weight':'int64'})
df
```

```python
df=df.drop('aircraft_version', axis=1)
df
```

```python
df.isnull().sum()
```

```python
df['year']=df['activity_period']/100
df
```

```python
df['month']=df['activity_period']%100
df
```

```python
df['year'] = df['year'].apply(np.int64)
df
```

```python
df=df.drop('activity_period', axis=1)
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```
